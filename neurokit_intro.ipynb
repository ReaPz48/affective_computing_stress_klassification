```python
#imports
import neurokit2 as nk
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sb
```

```python
#read polar h10 data textfile
array_polar = np.loadtxt("polar_sensordata/2023_08_06_23_I.txt", dtype=int)
```

```python
# Process ecg
ecg_signals, info = nk.ecg_process(array_polar[1000:1300], sampling_rate=130)
```

```python
# Extract R-peaks locations
_, rpeaks = nk.ecg_peaks(ecg_signals, sampling_rate=130)
```

```python
# plot ecg signal
nk.signal_plot(array_polar[610:800], sampling_rate=130)
```

```python
events = nk.events_find(array_polar[600:1100])
nk.events_plot(events, array_polar[600:1100])
```

```python
# Delineate the ECG signal and visualizing all peaks of ECG complexes
_, waves_peak = nk.ecg_delineate(ecg_signals, 
                                 rpeaks, 
                                 sampling_rate=130, 
                                 method="peak", 
                                 show=True, 
                                 show_type='peaks')
```

```python
# Process ecg
ecg_signals, info = nk.ecg_process(array_polar[10000:11000], sampling_rate=130)

nk.ecg_plot(ecg_signals)
```

```python
# Find peaks
peaks, info = nk.ecg_peaks(array_polar, sampling_rate=130)

# Compute HRV indices
hrv = nk.hrv_time(peaks, sampling_rate=130, show=True)
```
