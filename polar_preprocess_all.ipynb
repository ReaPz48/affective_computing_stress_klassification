```python
#imports
import neurokit2 as nk
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sb

epoch_step = 30 # in seconds
sample_rate = 130
sample_rate_seconds = 1/sample_rate
```

## Probanden

```python
probanden_dict = [{'id' : '2023_08_06_23_I.txt', 'save' : 'polar_I.csv', 'stress_begin' : 1220, 'stress_end' : 5651}, # isaac
                  {'id' : '2023-06-16_19-51-29_A.txt', 'save' : 'polar_A.csv', 'stress_begin' : 1600, 'stress_end' : 2880}, # anna
                  {'id' : '2023-06-16_20-45-59_P.txt', 'save' : 'polar_P.csv', 'stress_begin' : 660, 'stress_end' : 4800}, # peda probe
                  {'id' : '2023-06-17_19-19-19_PC.txt', 'save' : 'polar_PC.csv', 'stress_begin' : 900, 'stress_end' : 3000}, # peda konzert
                  {'id' : '2023-06-17_20-29-25_M.txt', 'save' : 'polar_M.csv', 'stress_begin' : 2400, 'stress_end' : 4560}, # marco
                  {'id' : '2023-06-23_14-47-56_C.txt', 'save' : 'polar_C.csv', 'stress_begin' : 1500, 'stress_end' : 2880} # christine
                  ]

probanden_dict
```

```python
for i in range(len(probanden_dict)):
    #read polar h10 data textfile
    array_polar = np.loadtxt("polar_sensordata/" + probanden_dict[i]['id'], dtype=int)
    #print(array_polar)

    stress_begin_row = int(probanden_dict[i]['stress_begin'] / epoch_step)
    stress_end_row = int(probanden_dict[i]['stress_end'] / epoch_step) - 1

    # Process ecg
    ecg_signals, info = nk.ecg_process(array_polar, sampling_rate=sample_rate)
    #print(ecg_signals)

    # segment the data
    epochs_polar = nk.epochs_create(ecg_signals, 
                            sampling_rate=sample_rate, 
                            epochs_start=0, 
                            epochs_end=epoch_step)
    
    #print(epochs_polar)

    # Analyze
    result_preprocess = nk.ecg_intervalrelated(epochs_polar)
    #print(result_preprocess)
    result_preprocess = result_preprocess.replace([np.inf, -np.inf], np.nan).dropna(axis=1)

    result_preprocess = result_preprocess.drop(columns=['Label'])
    polar_non_stress = result_preprocess[:stress_begin_row]  # First dataframe 
    polar_stress = result_preprocess[stress_begin_row:]  # Second dataframe

    polar_non_stress.insert(0, 'Stress', 0)
    polar_stress.insert(0, 'Stress', 1)

    result_polar = pd.concat([polar_non_stress, polar_stress], ignore_index=True)
    result_polar = result_polar.dropna(axis=1)

    # save to csv
    result_polar.to_csv('output_preprocessing/' + probanden_dict[i]['save'], index=False,)
    
    #array_polar.clear ()
    #ecg_signals.clear ()
    #info.clear ()
    #epochs_polar.clear ()
    #result_preprocess.clear ()
    #polar_non_stress.clear ()
    #polar_stress.clear ()
    #result_polar.clear ()
```
